### The 2-dimensional vectors are a = (0.5, 1.5), b = (4, 4), c = (8, 6), and x = (2, 2). Which of the three vectors a, b, and c is
(i) most similar to x according to dot product similarity?

(ii) most similar to x according to cosine similarity?

(iii) closest to x according to Euclidean distance?

```python
import numpy as np

a = np.array([0.5, 1.5])
b = np.array([4, 4])
c = np.array([8, 6])
x = np.array([2, 2])

vectors = {
    "a": a,
    "b": b,
    "c": c
}

# ==========================================
# (i) 内積類似性
# ==========================================

print("===== (i) 内積類似性 =====\n")

dot_results = {}

for name, vec in vectors.items():

    dot = np.dot(x, vec)
    dot_results[name] = dot

    print(f"x・{name} =", dot)

most_similar_dot = max(dot_results, key=dot_results.get)

print("\n最も類似しているベクトル =", most_similar_dot)

# ==========================================
# (ii) 余弦類似性
# ==========================================

print("\n\n===== (ii) 余弦類似性 =====\n")

cos_results = {}

for name, vec in vectors.items():

    cosine = np.dot(x, vec) / (np.linalg.norm(x) * np.linalg.norm(vec))

    cos_results[name] = cosine

    print(f"cos(x, {name}) =", cosine)

most_similar_cos = max(cos_results, key=cos_results.get)

print("\n最も類似しているベクトル =", most_similar_cos)

# ==========================================
# (iii) ユークリッド距離
# ==========================================

print("\n\n===== (iii) ユークリッド距離 =====\n")

dist_results = {}

for name, vec in vectors.items():

    distance = np.linalg.norm(x - vec)

    dist_results[name] = distance

    print(f"d(x, {name}) =", distance)

most_similar_dist = min(dist_results, key=dist_results.get)

print("\n最も類似しているベクトル =", most_similar_dist)
```
