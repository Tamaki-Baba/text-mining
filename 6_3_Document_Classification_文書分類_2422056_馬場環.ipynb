### Based on the data in Table,
(i) estimate a multinomial Naive Bayes classifier

(ii) apply the classifier to the test document

(iii) estimate a Bernoulli NB classifier

(iv) apply the classifier to the test document

```python
from collections import Counter
import math

train_data = [
    (["Taipei", "Taiwan"], "yes"),
    (["Macao", "Taiwan", "Shanghai"], "yes"),
    (["Japan", "Sapporo"], "no"),
    (["Sapporo", "Osaka", "Taiwan"], "no")
]

test_doc = ["Taiwan", "Taiwan", "Sapporo"]

vocab = set()

for words, label in train_data:
    vocab.update(words)

vocab = list(vocab)
V = len(vocab)

# (i) 多項ナイーブベイズ分類器を推定

print("===== (i) 多項ナイーブベイズ分類器 =====\n")

class_counts = Counter()

word_counts = {
    "yes": Counter(),
    "no": Counter()
}

total_words = {
    "yes": 0,
    "no": 0
}

# 単語頻度カウント
for words, label in train_data:

    class_counts[label] += 1

    for w in words:
        word_counts[label][w] += 1
        total_words[label] += 1

# 事前確率
P_yes = class_counts["yes"] / len(train_data)
P_no = class_counts["no"] / len(train_data)

print("P(yes) =", P_yes)
print("P(no) =", P_no)

print("\n--- 単語確率 P(w|class) ---")

for w in vocab:

    p_w_yes = (word_counts["yes"][w] + 1) / (total_words["yes"] + V)
    p_w_no = (word_counts["no"][w] + 1) / (total_words["no"] + V)

    print(f"{w:10s} | yes = {p_w_yes:.4f} | no = {p_w_no:.4f}")

# (ii) テスト文書へ適用

print("\n===== (ii) 多項ナイーブベイズの分類 =====\n")

score_yes = math.log(P_yes)
score_no = math.log(P_no)

for w in test_doc:

    p_w_yes = (word_counts["yes"][w] + 1) / (total_words["yes"] + V)
    p_w_no = (word_counts["no"][w] + 1) / (total_words["no"] + V)

    score_yes += math.log(p_w_yes)
    score_no += math.log(p_w_no)

print("Test document =", test_doc)

print("\nlog P(yes|d) =", score_yes)
print("log P(no|d)  =", score_no)

if score_yes > score_no:
    print("\n分類結果: China = yes")
else:
    print("\n分類結果: China = no")

# (iii) ベルヌーイナイーブベイズ分類器を推定

print("\n\n===== (iii) ベルヌーイナイーブベイズ分類器 =====\n")

docs_yes = [set(words) for words, label in train_data if label == "yes"]
docs_no = [set(words) for words, label in train_data if label == "no"]

n_yes = len(docs_yes)
n_no = len(docs_no)

print("--- 出現確率 P(w|class) ---")

for w in vocab:

    count_yes = sum(1 for d in docs_yes if w in d)
    count_no = sum(1 for d in docs_no if w in d)

    p_yes = (count_yes + 1) / (n_yes + 2)
    p_no = (count_no + 1) / (n_no + 2)

    print(f"{w:10s} | yes = {p_yes:.4f} | no = {p_no:.4f}")

# (iv) テスト文書へ適用

print("\n===== (iv) ベルヌーイナイーブベイズの分類 =====\n")

test_set = set(test_doc)

score_yes = math.log(P_yes)
score_no = math.log(P_no)

for w in vocab:

    count_yes = sum(1 for d in docs_yes if w in d)
    count_no = sum(1 for d in docs_no if w in d)

    p_yes = (count_yes + 1) / (n_yes + 2)
    p_no = (count_no + 1) / (n_no + 2)

    if w in test_set:
        score_yes += math.log(p_yes)
        score_no += math.log(p_no)
    else:
        score_yes += math.log(1 - p_yes)
        score_no += math.log(1 - p_no)

print("Test document =", test_doc)

print("\nlog P(yes|d) =", score_yes)
print("log P(no|d)  =", score_no)

if score_yes > score_no:
    print("\n分類結果: China = yes")
else:
    print("\n分類結果: China = no")
```
