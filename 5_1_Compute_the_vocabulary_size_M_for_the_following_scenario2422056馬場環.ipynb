### Looking at a collection of web pages, you find that there are 3000 different terms in the first 10,000 tokens and 30,000 different terms in the first 1,000,000 tokens.　Assume a search engine indexes a total of 20,000,000,000 (2 × 10^10) pages, containing 200 tokens/page on average.
What is the size of the vocabulary of the indexed collection as predicted by Heaps’ law?

```python
import math

# 与えられたデータ
V1 = 3000
n1 = 10000

V2 = 30000
n2 = 1000000

# Heaps' law
# V = K * n^β

# β を求める
beta = math.log(V2 / V1) / math.log(n2 / n1)

# K を求める
K = V1 / (n1 ** beta)

# 総トークン数
pages = 2 * 10**10
tokens_per_page = 200

n_total = pages * tokens_per_page

# 語彙サイズを予測
V_total = K * (n_total ** beta)

# 表示
print("beta =", beta)
print("K =", K)
print("Total tokens =", n_total)
print("Predicted vocabulary size =", V_total)
```
